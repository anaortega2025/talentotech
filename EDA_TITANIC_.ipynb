## Análisis bivariado

### Subtask:
Explorar las relaciones entre pares de variables, como la relación entre 'Survived' y otras variables ('Pclass', 'Sex', 'Age', 'Embarked', 'Fare').


**Reasoning**:
Create grouped bar plots to visualize the relationship between 'Survived' and categorical variables ('Pclass', 'Sex', 'Embarked') by calculating the proportion of survivors for each category, and create box plots to visualize the distribution of numerical variables ('Age', 'Fare') for survived and not survived groups.



```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Relationship between Survived and Pclass
pclass_survived = df.groupby('Pclass')['Survived'].value_counts(normalize=True).unstack()
pclass_survived.plot(kind='bar', stacked=False, ax=axes[0])
axes[0].set_title('Survival Rate by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Proportion')
axes[0].tick_params(axis='x', rotation=0)
axes[0].legend(title='Survived', labels=['No', 'Yes'])

# Relationship between Survived and Sex
sex_survived = df.groupby('Sex')['Survived'].value_counts(normalize=True).unstack()
sex_survived.plot(kind='bar', stacked=False, ax=axes[1])
axes[1].set_title('Survival Rate by Sex')
axes[1].set_xlabel('Sex')
axes[1].set_ylabel('Proportion')
axes[1].tick_params(axis='x', rotation=0)
axes[1].legend(title='Survived', labels=['No', 'Yes'])


# Relationship between Survived and Embarked
embarked_survived = df.groupby('Embarked')['Survived'].value_counts(normalize=True).unstack()
embarked_survived.plot(kind='bar', stacked=False, ax=axes[2])
axes[2].set_title('Survival Rate by Embarked')
axes[2].set_xlabel('Embarked')
axes[2].set_ylabel('Proportion')
axes[2].tick_params(axis='x', rotation=0)
axes[2].legend(title='Survived', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()

# Relationship between Survived and Age (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()

# Relationship between Survived and Fare (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

**Reasoning**:
Calculate and print the correlation matrix for the numerical features and then create a heatmap to visualize the correlation matrix.



```python
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare', 'Survived']
correlation_matrix = df[numerical_features].corr()
print("Correlation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   **Survival Rate:** A smaller proportion of passengers survived compared to those who did not survive.
*   **Passenger Class Distribution:** The majority of passengers were in Pclass 3 (55\%), followed by Pclass 1 (24\%) and Pclass 2 (21\%).
*   **Gender Distribution:** There were significantly more male passengers (64.8\%) than female passengers (35.2\%).
*   **Age Distribution:** The age distribution is slightly right-skewed, with a peak around 20-30 years old and some outliers at older ages. The median age for both survived and not survived groups is similar.
*   **Family Size:** Most passengers traveled without siblings or spouses (SibSp=0), and without parents or children (Parch=0).
*   **Fare Distribution:** The fare distribution is highly right-skewed, with most passengers paying lower fares and some outliers paying very high fares. Passengers who survived generally paid higher fares than those who did not survive.
*   **Embarkation Port:** The majority of passengers embarked from Southampton (S) (72.4\%), followed by Cherbourg (C) (18.9\%) and Queenstown (Q) (8.7\%). Passengers who embarked at Cherbourg (C) had a higher survival rate compared to those from Southampton (S) and Queenstown (Q).
*   **Survival by Pclass:** Passengers in higher classes (1 and 2) had a much higher survival rate than those in Pclass 3.
*   **Survival by Sex:** Female passengers had a significantly higher survival rate than male passengers.
*   **Correlation with Survived:** 'Fare' shows a moderate positive correlation with 'Survived' (0.26). 'Age', 'SibSp', and 'Parch' show weak correlations with 'Survived'.
*   **Survival by Pclass and Sex:** Within each Pclass, females consistently had a higher survival rate than males. The survival rate for females in Pclass 1 was particularly high.

### Insights or Next Steps
