```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
try:
    dataset = pd.read_csv('Salary_Data.csv')
except FileNotFoundError:
    print("Salary_Data.csv not found. Using a sample dataset as a placeholder.")
    # As a placeholder, you might want to create a dummy DataFrame or load a sample dataset
    # For now, let's assume the file is available or create a simple one for demonstration
    # Example of creating a dummy DataFrame:
    data = {'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5],
            'Salary': [39343.0, 46205.0, 37731.0, 43525.0, 39891.0, 56642.0, 60150.0, 54445.0, 64445.0, 57189.0, 63218.0, 55794.0, 56957.0, 57081.0, 61111.0, 67938.0, 66029.0, 83000.0, 81363.0, 93940.0, 91738.0, 98273.0, 101000.0, 113812.0, 109431.0, 105582.0, 116969.0, 112635.0, 122391.0, 121872.0]}
    dataset = pd.DataFrame(data)


X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values # Assuming the last column is the target variable
```

```python
# Splitting the dataset into the Training set and Test set /Dividiendo el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set/Ajuste de regresión lineal simple al conjunto de entrenamiento
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results/Prediciendo los resultados del conjunto de prueba
y_pred = regressor.predict(X_test)
```

```python
# Evaluating the Model with Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
```

```python
# Visualising the Test set results/Visualizando los resultados del conjunto de prueba
plt.scatter(X_test.flatten(), y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Plotting a histogram of actual test set salaries
plt.hist(y_test, bins=10, alpha=0.7, label='Actual Salaries')
plt.title('Histogram of Actual Test Set Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Plotting a histogram of predicted test set salaries
plt.hist(y_pred, bins=10, alpha=0.7, label='Predicted Salaries', color='orange')
plt.title('Histogram of Predicted Test Set Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Calculating residuals
residuals = y_test - y_pred

# Plotting a histogram of residuals
plt.hist(residuals, bins=10)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals (Actual - Predicted)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculating Mean Absolute Percentage Error (MAPE)
# Avoid division by zero by adding a small epsilon or filtering out zero actual values
# Here, we'll filter out actual values of zero if they exist
y_test_filtered = y_test[y_test != 0]
y_pred_filtered = y_pred[y_test != 0]

mape = np.mean(np.abs((y_test_filtered - y_pred_filtered) / y_test_filtered)) * 100

print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
```
