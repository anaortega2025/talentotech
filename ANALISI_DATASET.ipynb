## Análisis univariado

### Subtask:
Analizar la distribución de cada variable individualmente (histogramas, box plots, conteos de valores únicos).


**Reasoning**:
Generate histograms and box plots for all numerical columns and display value counts for any categorical columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Get numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms for numerical columns
print("Histograms for Numerical Columns:")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot box plots for numerical columns
print("\nBox Plots for Numerical Columns:")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()

# Get categorical columns (if any)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Display value counts for categorical columns
if len(categorical_cols) > 0:
    print("\nValue Counts for Categorical Columns:")
    for col in categorical_cols:
        print(f"\nValue counts for {col}:")
        display(df[col].value_counts())
else:
    print("\nNo categorical columns found in the DataFrame.")
```

## Análisis bivariado

### Subtask:
Explorar la relación entre pares de variables (gráficos de dispersión, matrices de correlación, tablas cruzadas).


**Reasoning**:
Calculate and display the correlation matrix for the numerical columns and create a heatmap to visualize the correlations.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculate the correlation matrix for the numerical columns
correlation_matrix = df.corr()

# 2. Display the correlation matrix
print("Correlation Matrix:")
display(correlation_matrix)

# 3. Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

**Reasoning**:
Based on the correlation matrix and heatmap, select a few pairs of variables with notable correlations and create scatter plots to visualize these relationships. Pairs with strong positive correlations (e.g., total_rooms and total_bedrooms, population and households) and a strong negative correlation (longitude and latitude) are good candidates. Also, include pairs with moderate correlation with the target variable (median_house_value and median_income).



```python
# 4. Create scatter plots for selected pairs of numerical variables

# Select pairs with strong correlations
strong_positive_pairs = [('total_rooms', 'total_bedrooms'), ('population', 'households')]
strong_negative_pair = [('longitude', 'latitude')]
moderate_target_correlation_pairs = [('median_income', 'median_house_value')]

selected_pairs = strong_positive_pairs + strong_negative_pair + moderate_target_correlation_pairs

print("\nScatter Plots for Selected Pairs of Numerical Features:")

for x_col, y_col in selected_pairs:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=x_col, y=y_col, alpha=0.6)
    plt.title(f'Scatter Plot of {x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()
```

## Visualización de datos

### Subtask:
Crear visualizaciones relevantes para entender mejor los patrones y relaciones en los datos.


**Reasoning**:
Create the requested visualizations to understand patterns and relationships in the data.



```python
# 2. Create a histogram for the median_house_value column
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], kde=True, bins=50)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()

# 3. Create a box plot for the median_house_value column
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['median_house_value'])
plt.title('Box Plot of Median House Value')
plt.ylabel('Median House Value')
plt.show()

# 4. Create a scatter plot of median_income vs. median_house_value
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='median_income', y='median_house_value', alpha=0.6)
plt.title('Scatter Plot of Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()

# 5. Create a scatter plot of longitude vs. latitude, colored by median_house_value
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='median_house_value', palette='viridis', alpha=0.6, s=10)
plt.title('Geographical Distribution of Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
